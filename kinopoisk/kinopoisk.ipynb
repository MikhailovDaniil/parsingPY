### использование библиотеки requests для получения html вёрстки сайта

```python
import requests
```

```python
url = "https://www.kinopoisk.ru/lists/top250/"

r = requests.get(url)
r.text
```

### использование библиотеки bs4 для работы с полученными данными

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text, "lxml")
soup
```

#### для формирования списка фильмов нам требуется извлекать информацию из тега <br>
<font color="green"> div class="desktop-rating-selection-film-item" </font>

```python
soup.find("div", class_ ="desktop-rating-selection-film-item")
```

#### получим ссылку на каждый фильм (она находится в теге "a" с классом "selection-film-item-meta__link")

```python
soup.find("div", class_ ="desktop-rating-selection-film-item").find("a", class_="selection-film-item-meta__link")
```

```python
soup.find("div",
          class_ ="desktop-rating-selection-film-item").find("a",
                                                             class_="selection-film-item-meta__link").get("href")
```

```python
link = "kinopoisk.ru"+ soup.find("div",
          class_ ="desktop-rating-selection-film-item").find("a",
                                                             class_="selection-film-item-meta__link").get("href")

link
```

#### получим название фильма на русском

```python
soup.find("div",
          class_ ="desktop-rating-selection-film-item").find("p",
                                                             class_="selection-film-item-meta__name").text
```
